```python
# Load Libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# initial set up
url = "https://www.cars24.com/buy-used-car?sort=bestmatch&serveWarrantyCount=true&listingSource=ViewAllCars&storeCityId=4709"
region_urls = []

# Read HTML
response = requests.get(url)
content  = BeautifulSoup(response.content, 'html.parser')

# Get URLs for each Region
data    = content.find('div', class_='_3t7GZ')
regions = data.find_all('a')
for region in regions: 
    region_urls.append(region.attrs['href'])
```

```python

# Save Region Urls as pickle
regions_cars24 = pd.DataFrame({'RegionURL':region_urls})
regions_cars24
```

```python
# Save Pickle file
regions_cars24.to_pickle('regions_cars24.pkl')
```
